```python
import pandas as pd
import numpy as np
import matplotlib as plm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dati_J48 = pd.read_csv("dati_J48.csv")
dati_RF = pd.read_csv("dati_RF.csv")
dati_LR = pd.read_csv("dati_LR.csv")
dati_SVM = pd.read_csv("dati_SVM.csv")
dati_MLP = pd.read_csv("dati_MLP.csv")
dati_KNN = pd.read_csv("dati_KNN.csv")
```

```python
dati_KNN
```

```python
fig1 = plt.figure()
plt.style.use('seaborn-whitegrid')
plt.plot(dati_J48['minNumObj'], dati_J48['Objective value'], )
plt.title("J48 parameter optimization")
plt.xlabel('minNumObj')
plt.ylabel('Cost')
fig1.savefig('J48.png')
```

```python
fig2 = plt.figure()
plt.style.use('seaborn-whitegrid')
plt.plot(dati_RF['maxDepth'], dati_RF['Objective value'], )
plt.title("RF parameter optimization")
plt.xlabel('maxDepth')
plt.ylabel('Cost')
fig2.savefig('RF.png')
```

```python
fig3 = plt.figure()
plt.style.use('seaborn-whitegrid')
plt.plot(dati_LR['ridge'], dati_LR['Objective value'], )
plt.title("LR parameter optimization")
plt.xlabel('ridge')
plt.ylabel('Cost')
fig3.savefig('LR.png')
```

```python
fig4 = plt.figure()
plt.style.use('seaborn-whitegrid')
plt.plot(dati_SVM['c'], dati_SVM['Objective value'], )
plt.title("SVM parameter optimization")
plt.xlabel('complexity')
plt.ylabel('Cost')
fig4.savefig('SVM.png')
```

```python
fig5 = plt.figure()
plt.style.use('seaborn-whitegrid')
plt.plot(dati_MLP['hidden_layer'], dati_MLP['Objective value'], )
plt.title("MLP parameter optimization")
plt.xlabel('hidden_layer')
plt.ylabel('Cost')
fig5.savefig('MLP.png')
```

```python
fig6 = plt.figure()
plt.style.use('seaborn-whitegrid')
plt.plot(dati_KNN['KNN'], dati_KNN['Objective value'], )
plt.title("KNN parameter optimization")
plt.xlabel('K')
plt.ylabel('Cost')
fig6.savefig('KNN.png')
```
